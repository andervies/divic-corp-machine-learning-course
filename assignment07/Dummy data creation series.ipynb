## Problem One: Creating random numbers

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)

mean = (-3, 0)
cov = [[1.0, 0.8], [0.8, 1.0]]


rand_num = np.random.multivariate_normal(mean, cov, size=500)
rand_num.shape
```

## Problem Two: Visualisation by scatter plot

```python
plt.title("Visualisation by Scatter Plot")
plt.xlabel("x1")
plt.ylabel("x2")
plt.scatter(rand_num[:, 0], rand_num[:, 1])
plt.show()
```

## Problem Three: Visualisation by Histogram

```python
plt.title("Histogram Showing x1 data", c="navy", fontsize=14)
plt.xlabel("x1 values", c="grey")
plt.ylabel("Frequency", c="grey")
plt.hist(rand_num[:, 0], bins=50)
plt.xlim(left=-5.57)
plt.show()
```

```python
plt.title("Histogram Showing x2 data", c="maroon", fontsize=14)
plt.xlabel("x2 values", c="grey")
plt.ylabel("Frequency", c="grey")
plt.hist(rand_num[:, 1], bins=50)
plt.xlim(right=2.7)
plt.show()
```

## Problem Four: Addition of data

```python
new_mean = (0, -3)

np.random.seed(0)

rand_num2 = np.random.multivariate_normal(new_mean, cov, 500)
rand_num2.shape
```

```python
plt.title("Visualisation by Scatter Plot")
plt.xlabel("x1")
plt.ylabel("x2")
plt.scatter(rand_num[:, 0], rand_num[:, 1])


plt.scatter(rand_num2[:, 0], rand_num2[:, 1])
plt.legend(labels=[0,1])
plt.show()
```

## Problem Five: Data combination

```python
combined_arrays = np.concatenate((rand_num, rand_num2))
```

```python
combined_arrays.shape
```

## Problem Six: Labeling

```python
zeros = np.zeros(500)
ones = np.ones(500)

question_labels = np.concatenate((zeros, ones))
combined_arrays2 = np.hstack((combined_arrays, question_labels[:, None]))
combined_arrays2.shape
```
