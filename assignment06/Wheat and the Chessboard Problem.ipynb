## Problem Two: Expansion to n × m mass

```python
# Creating function to return an ndarray containing the wheats on an nxm chessboard using a for loop and python list
def wheat_by_chessboard_dim(n_dim, m_dim):
    num_of_squares = n_dim * m_dim
    board_list = [1]
    for _ in range(num_of_squares - 1):
        board_list.append(2*board_list[-1])
    board_ndarray = np.array(board_list)
    return board_ndarray.reshape(n_dim, m_dim)


# Finding the number of wheats on an 8x8 square chessboard
wheats_for_8x8 = wheat_by_chessboard_dim(8,8)
print(f"The number of wheats on each square of an 8x8 square chessboard is \n{wheats_for_8x8}")
```

```python
# Finding the average of each column

column_averages = []
for col in range(8):
    current_col_average = wheats_for_8x8[col].mean()
    column_averages.append(current_col_average)
    
column_averages
```

## Problem Six: Expansion to n × m mass by another calculation method

```python
# Creating same function from problem 2, this time using np.append()
def wheat_by_chessboard_dim(n_dim, m_dim):
    num_of_squares = n_dim * m_dim
    board_ndarray = np.array([1]).astype(np.uint64)
    for _ in range(num_of_squares - 1):
        board_ndarray = np.append(board_ndarray, 2*board_ndarray[-1])
    return board_ndarray.reshape(n_dim, m_dim)


# Finding the number of wheats on an 8x8 square chessboard
print(f"The number of wheats on each square of an 8x8 square chessboard is \n{wheat_by_chessboard_dim(8,8)}")
```

```python
# Creating same function from problem 2, this time using broadcast
def wheat_by_chessboard_dim(n_dim, m_dim):
    num_of_squares = n_dim * m_dim
    indices = np.arange(num_of_squares).astype(np.uint64)
    all_wheats = 2**indices
    return all_wheats.reshape(n_dim, m_dim)


# Finding the number of wheats on an 8x8 square chessboard
print(f"The number of wheats on each square of an 8x8 square chessboard is \n{wheat_by_chessboard_dim(8,8)}")
```

## Problem Seven: Comparing calculation times

```python
%%timeit
# timing the first function which uses the python list and a for loop


def wheat_by_chessboard_dim(n_dim, m_dim):
    num_of_squares = n_dim * m_dim
    board_list = [1]
    for _ in range(num_of_squares - 1):
        board_list.append(2*board_list[-1])
    board_ndarray = np.array(board_list)
    return board_ndarray.reshape(n_dim, m_dim)


wheat_by_chessboard_dim(8,8)
```

```python
%%timeit
# timing the second function which uses the np.array(), np.append(), and a for loop


def wheat_by_chessboard_dim(n_dim, m_dim):
    num_of_squares = n_dim * m_dim
    board_ndarray = np.array([1]).astype(np.uint64)
    for _ in range(num_of_squares - 1):
        board_ndarray = np.append(board_ndarray, 2*board_ndarray[-1])
    return board_ndarray.reshape(n_dim, m_dim)


wheat_by_chessboard_dim(8,8)
```

```python
%%timeit
# timing the third function which uses the broadcasting method


def wheat_by_chessboard_dim(n_dim, m_dim):
    num_of_squares = n_dim * m_dim
    indices = np.arange(num_of_squares).astype(np.uint64)
    all_wheats = 2**indices
    return all_wheats.reshape(n_dim, m_dim)


wheat_by_chessboard_dim(8,8)
```

#### Comparing the timeit outputs of the three functions

With an average execution time of about 3.57 micro seconds after 7 runs, the third function appears to be the most time efficient. This result is expected because it does not use a for loop, it instead leverages on vector operations(broadcasting in this case) which numpy is optimised for.

The difference in the run times of the first and second functions turned out to be a little shocking. The function using the in-built python list appears to be almost 23 times faster than the function that calls np.append() in the for loop. This result can be understood in the fact that in the second funtion, numpy creates a new array in every loop and copies the existing elements, then adds a new one. This approach seems less time efficient than using the python list and then converting the entire list into a numpy array.
