# Problem 1
##  Number of grains of rice on the 100th day

```python
# Calculating the total number of grains to be recieved on the 100th day

grains_per_day = []

for day in range(0, 100):
    if day == 0:
        total_grains = 1
        grains_per_day.append(total_grains)
    else:
        total_grains *= 2
        grains_per_day.append(total_grains)
print(f"The total number of grains by the 100th day is: {sum(grains_per_day) :,}")
```

```python
# Plotting a line graph representing the change in number of grains with respect to day it was was recieved

import matplotlib.pyplot as plt
%matplotlib inline

plt.title("Increment in Grains vs Day", color="blue", fontsize=20)
plt.xlabel("Days", color="teal", fontsize=15)
plt.ylabel("Rice Grains", color="teal", fontsize=15)
plt.plot(range(1,101), grains_per_day, color="darkblue")
```

# Problem 2
## Number of rice grains outside of the 100th day

```python
# Defining the days for the duration of the course assuming the course is taken 3 times a week for 52 weeks
COURSE_DAYS = 3 * 52


def compute_sorori_shinzaemon(n_days=100):
    """A function that calculates the number of rice grains per day in the rice anecdote of Shinzaemon Sorori.

    Parameteres
    -----------
    n_days: int
        Number of days to get rice (default : 100)

    Returns
    -------
    list_n_grains : list
        List of the number of rice you get one day (list length equals n_days)
    list_total_grains : list
        List of total number of rice received by a certain day (list length is equal to n_days)
    """

    grains_per_day = []
    list_n_grains = []
    for day in range(1, n_days+1):
        list_n_grains.append(day)
        if day == 1:
            total_grains = 1
            grains_per_day.append(total_grains)
        else:
            total_grains *= 2
            grains_per_day.append(total_grains)

    return list_n_grains, grains_per_day


# Computing the number of grains recievable for the number of days I shall be taking this course

all_course_days, grains_per_course_day = compute_sorori_shinzaemon(n_days=COURSE_DAYS)
```

```python
# Line graph showing the recievable grains for each day I shall be taking this course


plt.title("Increment in Grains for Duration of Course", color="darkblue", fontsize=14)
plt.xlabel("Days", color="darkred", fontsize=12)
plt.ylabel("Rice Grains", color="darkgreen", fontsize=12)
plt.plot(all_course_days, grains_per_course_day, color="teal", linewidth=2)
```

# Problem 3
##  How many people can live for how many days with the rice we get?

For this problem, I googled the average number of rice grains per plate, and found, according to a post on https://fanaticallyfood.com, that a standard bowl of rice would contain an average of 3000 grains. Then I googled the recommended number of meals per individual per day, the concensus is three meals per day. I used these results in this problem.

```python
# A function to calculate the days it would take a certain amount of people to live on a certain amount of grains

grains_per_bowl = 3000
grains_per_3_meals = grains_per_bowl * 3


def compute_days_to_live(people_num, grains):
    grains_per_people_day = grains_per_3_meals * people_num
    days_to_live = grains // grains_per_people_day
    return days_to_live
```

```python
# Computing the number of days the class can survive on the grains recievable for the duration of the course
# Here, I assume the number of students in the class is 250

STUDENTS = 250

print(f"The number of days it would take {STUDENTS} to survive on {sum(grains_per_course_day) :,} \
\ngrains is: {compute_days_to_live(STUDENTS, sum(grains_per_course_day)) :,}")
```
